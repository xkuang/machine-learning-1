```python
import pandas as pd
filmlistesi = pd.read_csv("http://www.datascience.istanbul/data/imdbratings.data")
```

```python
filmlistesi.shape
```

```python
filmlistesi.columns
```

```python
filmUzunlukKategori =[]
for i in filmlistesi.duration:
    if i > 100:
        filmUzunlukKategori.append('çokUzun')
    elif i >60 and i < 80:
        filmUzunlukKategori.append('orta')
    else:
        filmUzunlukKategori.append('kısa')
```

```python
filmlistesi["uzunluk"]=filmUzunlukKategori
```

```python
filmlistesi.head
```

```python
filmlistesi[filmlistesi.uzunluk=="çokUzun"]
```

```python
filmlistesi.groupby("content_rating").star_rating.mean().sort_values(ascending=False)
```

```python
%matplotlib inline
filmlistesi.groupby("content_rating").star_rating.mean().sort_values(ascending=False).plot(kind='bar')
```

```python
%matplotlib inline
filmlistesi.groupby("content_rating").star_rating.agg(['mean','count','max','min']).sort_values('mean',ascending=False).plot(kind='bar')
```
